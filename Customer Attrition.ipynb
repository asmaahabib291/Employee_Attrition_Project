### Classifying Employees By Tenure

```python
def classify_tenure(row):
    
    if row['YearsAtCompany'] < 1 and row['TotalWorkingYears'] <= 1:
        return 'Newbie'
    
    elif row['TotalWorkingYears'] <= 3 or row['YearsAtCompany'] < 2:
        return 'Junior'
    
    elif 3 < row['TotalWorkingYears'] <= 8 or 2 <= row['YearsAtCompany'] < 5:
        return 'Mid'
    
    elif 8 < row['TotalWorkingYears'] <= 15 or 5 <= row['YearsAtCompany'] < 10:
        return 'Senior'
    
    elif row['TotalWorkingYears'] > 15 and row['YearsAtCompany'] >= 10:
        return 'Expert'
    
    else:
        return 'Mid'


df['TenureGroup'] = df.apply(classify_tenure, axis=1)

```

```python
import pandas as pd
from scipy.stats import ttest_ind

df = pd.read_csv("HR-Employee-Attrition-Cleaned.csv")

if df["Attrition"].dtype == object:
    df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

numerical_cols = [col for col in numerical_cols if col != "Attrition"]

results = []

for col in numerical_cols:
    group_yes = df[df["Attrition"] == 1][col]
    group_no = df[df["Attrition"] == 0][col]
    
    t_stat, p_val = ttest_ind(group_yes, group_no, equal_var=False)
    
    results.append({
        "Feature": col,
        "T-Statistic": round(t_stat, 3),
        "P-Value": round(p_val, 5)
    })

ttest_df = pd.DataFrame(results).sort_values(by="P-Value")

ttest_df
```

```python

numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

exclude_columns = ['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber']
numeric_cols = [col for col in numeric_cols if col not in exclude_columns]
categorical_cols = [col for col in categorical_cols if col not in exclude_columns]

app = dash.Dash(__name__)
server = app.server  

app.layout = html.Div([
    html.H1("HR Employee Attrition Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Department"),
        dcc.Dropdown(
            id='department-filter',
            options=[{'label': dept, 'value': dept} for dept in df['Department'].dropna().unique()] + [{'label': 'All', 'value': 'All'}],
            value='All'
        )
    ], style={'width': '30%', 'display': 'inline-block', 'margin': '10px'}),

    html.Div([
        html.Label("Select Gender"),
        dcc.Dropdown(
            id='gender-filter',
            options=[{'label': g, 'value': g} for g in df['Gender'].dropna().unique()] + [{'label': 'All', 'value': 'All'}],
            value='All'
        )
    ], style={'width': '30%', 'display': 'inline-block', 'margin': '10px'}),

    html.Div([
        html.Label("Select Attrition Status"),
        dcc.Dropdown(
            id='attrition-filter',
            options=[{'label': str(a), 'value': a} for a in df['Attrition'].dropna().unique()] + [{'label': 'All', 'value': 'All'}],
            value='All'
        )
    ], style={'width': '30%', 'display': 'inline-block', 'margin': '10px'}),

    html.Div([
        html.Label("X-axis Variable"),
        dcc.Dropdown(
            id='x-axis',
            options=[{'label': col, 'value': col} for col in numeric_cols],
            value=numeric_cols[0] if numeric_cols else None
        )
    ], style={'width': '45%', 'display': 'inline-block', 'margin': '10px'}),

    html.Div([
        html.Label("Y-axis Variable"),
        dcc.Dropdown(
            id='y-axis',
            options=[{'label': col, 'value': col} for col in numeric_cols],
            value=numeric_cols[1] if len(numeric_cols) > 1 else numeric_cols[0]
        )
    ], style={'width': '45%', 'display': 'inline-block', 'margin': '10px'}),

    html.Div([
        html.Label("Color By"),
        dcc.Dropdown(
            id='color-by',
            options=[{'label': col, 'value': col} for col in categorical_cols],
            value='Attrition' if 'Attrition' in categorical_cols else categorical_cols[0]
        )
    ], style={'width': '45%', 'display': 'inline-block', 'margin': '10px'}),

    html.Div(id='summary-stats', style={'textAlign': 'center', 'fontSize': 18, 'margin': '20px'}),

    dcc.Graph(id='scatter-plot'),
    dcc.Graph(id='box-plot')
])

@app.callback(
    [Output('scatter-plot', 'figure'),
     Output('box-plot', 'figure'),
     Output('summary-stats', 'children')],
    [Input('department-filter', 'value'),
     Input('gender-filter', 'value'),
     Input('attrition-filter', 'value'),
     Input('x-axis', 'value'),
     Input('y-axis', 'value'),
     Input('color-by', 'value')]
)
def update_dashboard(department, gender, attrition, x_axis, y_axis, color_by):
    dff = df.copy()

    # Apply filters
    if department != 'All':
        dff = dff[dff['Department'] == department]
    if gender != 'All':
        dff = dff[dff['Gender'] == gender]
    if attrition != 'All':
        dff = dff[dff['Attrition'] == attrition]

    if dff.empty:
        return {}, {}, "⚠️ No data available for the selected filters."

    if len({x_axis, y_axis, color_by}) < 3:
        return {}, {}, "⚠️ Please select different variables for X, Y, and Color."

    total_employees = len(dff)

    if dff['Attrition'].dtype in ['int64', 'int32', 'float']:
        attrition_count = dff[dff['Attrition'] == 1].shape[0]
    else:
        attrition_count = dff[dff['Attrition'] == 'Yes'].shape[0]

    attrition_rate = round((attrition_count / total_employees) * 100, 2)

    summary = f"Total Employees: {total_employees} | Attrition Count: {attrition_count} | Attrition Rate: {attrition_rate}%"

    fig_scatter = px.scatter(
        dff, x=x_axis, y=y_axis, color=color_by,
        title=f"{x_axis} vs {y_axis} colored by {color_by}"
    )

    fig_box = px.box(
        dff, x=color_by, y=y_axis, color=color_by,
        title=f"Distribution of {y_axis} by {color_by}"
    )

    return fig_scatter, fig_box, summary

if __name__ == '__main__':
    app.run_server(debug=True, port=8050)
```

## PipeLine Export

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
import joblib
```

```python

numerical_cols = ['Age','DistanceFromHome','MonthlyIncome','StockOptionLevel',
                  'RelationshipSatisfaction','YearsAtCompany','MonthlyRate',
                  'YearsSinceLastPromotion','TrainingTimesLastYear',
                  'JobSatisfaction','JobInvolvement','PerformanceRating']
categorical_cols = ['Gender','OverTime','BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
```

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
])

model = AdaBoostClassifier(DecisionTreeClassifier(max_depth=9),
                           learning_rate=0.99999, n_estimators=70)
```

```python

pipeline = ImbPipeline(steps=[
    ('preprocessing', preprocessor),
    ('smote', SMOTE(random_state=42)),
    ('model', model)
])
```

```python

pipeline.fit(X_train, y_train)
```

```python

joblib.dump(pipeline, "attrition_pipeline.pkl")
```
